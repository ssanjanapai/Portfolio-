```python
#Populate Property Address data

prop_addr=sqldf("Select * From housing_data order by ParcelID;")
prop_addr.head()
```

```python
prop_addr1=sqldf("Select a.ParcelID, a.PropertyAddress, b.ParcelID, b.PropertyAddress, IFNULL(a.PropertyAddress,b.PropertyAddress) as replacement From housing_data a JOIN housing_data b on a.ParcelID = b.ParcelID AND a.[UniqueID ] <> b.[UniqueID ] Where a.PropertyAddress is null;")
prop_addr1.head()
```

```python
housing_data['PropertyAddress'] = housing_data['PropertyAddress'].str.replace("None", str(prop_addr1['replacement']))
```
