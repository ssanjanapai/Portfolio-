```python
# joining the two tables
deaths_vaccinations= sqldf("Select * \
FROM data_deaths d join data_vaccinations v \
on d.location=v.location and d.date=v.date;")
```

```python
# Total population vs vaccination
pop_vacc = sqldf("SELECT d.continent,d.location,d.date,d.population,v.new_vaccinations FROM data_deaths d join data_vaccinations v ON d.location=v.location and d.date=v.date where d.continent is not null ORDER BY 1,2,3;")
```

```python
pop_vacc
```

```python
# Population vs vaccination
pop_vacc2= sqldf("select d.continent,d.location,d.date,d.population,v.new_vaccinations,SUM(cast(v.new_vaccinations as int)) OVER (PARTITION BY d.location ORDER BY d.location and d.date) as Rolling_Vaccination from data_deaths d join data_vaccinations v on d.location=v.location and d.date=v.date where d.continent is not null order by 2,3;")
```

```python
pop_vacc2
```

```python
pop_vacc_CTE = sqldf("""
    with PopvsVac as (
        select 
            d.continent, 
            d.location, 
            d.date, 
            d.population, 
            v.new_vaccinations, 
            SUM(cast(v.new_vaccinations as int)) OVER (PARTITION BY d.location ORDER BY d.location, d.date) as Rolling_Vaccination 
        from data_deaths d 
        join data_vaccinations v on d.location=v.location and d.date=v.date 
        where d.continent is not null 
        order by 2,3
    )
    select * from PopvsVac;
""")
```
